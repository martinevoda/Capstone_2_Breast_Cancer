## Introduction

The dataset is structured with 11 variables for each patient:

* pid: Patient identifier, a unique identifier assigned to each patient in the study.
* age: Age of the patient in years (21-80).
* meno: Menopausal status of the patient. It is represented as 0 for premenopausal and 1 for postmenopausal (0-1).
* size: Tumor size in millimeters (3-120).
* grade: Tumor grade, which provides information about the aggressiveness of the tumor (1-3).
* nodes: Number of positive lymph nodes, indicating the extent of lymph node involvement (1-51).
* pgr: Progesterone receptors measured in fmol/l (femtomoles per liter) (0-2380).
* er: Estrogen receptors measured in fmol/l (0-1144).
* hormon: Hormonal therapy given to the patient. It is represented as 0 for no hormonal therapy and 1 for receiving hormonal therapy (0-1).
* rfstime: Recurrence-free survival time in days, which refers to the duration until the first occurrence of recurrence, death, or the last follow-up (8-2659).
* status: Patient status indicator, where 0 represents being alive without recurrence, and 1 indicates recurrence or death (0-1).


The dataset contains patient records from a clinical trial conducted by the German Breast Cancer Study Group (GBSG) between 1984 and 1989.

```python
breast_cancer_2['status_label'] = breast_cancer_2['status']#.map({0: "alive without recurrence",1:"recurrence or death"})
plot = sns.catplot(x='status_label',data=breast_cancer_2,kind='count', palette='YlOrBr')
total=len(breast_cancer_2['status'])
for p in plot.ax.patches:
    percentage = '{:.1f}%'.format(100 * p.get_height()/total)
    x = p.get_x() + p.get_width() / 2
    y = p.get_height()
    plot.ax.text(x,y, percentage, ha='center',va='bottom')
plt.xlabel('Status')
plt.ylabel('Count')
plt.title('Distribution of Patient Stauts')
plt.show()
```

Upon examining the distribution of patient status, the following insights emerge:

Alive without Recurrence: Accounting for 56.4% of patients, are currently alive without experiencing a recurrence.
    
Recurrence or Death: The remaining 43.6% of patients have unfortunately faced either a recurrence of the condition or succumbed to it.

```python
plt.scatter(breast_cancer_2['size'],breast_cancer_2['nodes'], c="orange")
plt.xlabel('Size')
plt.ylabel('Node')
plt.title('Scatter plot of Size vs Nodes')
plt.show()
```

The scatter plot illustrates some pattern: as the tumor size (X-axis) increases, there is a tendency for the number of nodes (Y-axis) to increase as well.
    
The positive slope of the trendline or the general upward trend in the data points indicates a positive correlation between tumor size and the number of nodes. This suggests that larger tumors tend to have a higher number of associated nodes.
    
There are some outliers in the data. These data points deviate from the overall trend, indicating instances where the number of nodes may not strictly follow the expected pattern based on tumor size. Investigating these outliers can provide insights into potential anomalies or specific cases that warrant further examination.

Understanding the relationship between tumor size and node count is crucial for medical decision-making. It informs clinicians about the potential implications of tumor size on the spread or severity of the cancer.


```python
plt.figure(figsize=(8,6))
sns.boxplot(x='status',y='size',data=breast_cancer_2, palette="YlOrBr")
plt.xlabel('Status')
plt.ylabel('Tumor size (mm)')
plt.title('Status vs Tumor size(mm)')
plt.show()
```

The box plot depicts the distribution of 'Tumor Size' with respect to the 'Status' categories: 0 for "Alive without Recurrence" and 1 for "Recurrence or Death." Here are key observations:


Both 'Alive without Recurrence' (Status 0) and 'Recurrence or Death' (Status 1) categories exhibit boxes of similar height. This suggests comparable interquartile ranges and overall spread of tumor sizes within each status group.
Notably, within the 'Recurrence or Death' category (Status 1), there is an elevated box height between 35 and 40 on the y-axis. This indicates a concentration of tumor sizes within this range, suggesting a potential clustering of cases with specific tumor size characteristics.

The median line inside the box for 'Recurrence or Death' (Status 1) is slightly higher, around 30 on the y-axis, compared to the median of 'Alive without Recurrence' (Status 0). This indicates that, on average, tumor sizes in the 'Recurrence or Death' group may be somewhat larger.


Both 'Alive without Recurrence' and 'Recurrence or Death' status categories exhibit whiskers, suggesting variability in tumor sizes within each group.
Notably, the whiskers for the 'Recurrence or Death' category (Status 1) appear longer than those for 'Alive without Recurrence' (Status 0). This suggests a wider range of tumor sizes and potentially more extreme values within the 'Recurrence or Death' group.

Further statistical analysis may be warranted to confirm the significance of these observed patterns.

```python
def box_plot_features(data):
    data_to_plot = data.iloc[:, :-1]
    plt.figure(figsize=(10,6))
    plt.suptitle("Box Plots for Features in Breast Cancer Dataset: Exploring the Distribution and Outliers", fontsize=12)
    for i,column in enumerate(data_to_plot.columns):
        plt.subplot(1,len(data_to_plot.columns),i+1)
        sns.boxplot(data=data_to_plot[column],palette='YlOrBr',width=5.0)
        plt.title(column)
        plt.tight_layout()
        plt.show

box_plot_features(breast_cancer_2)
```

Observation 1: Notably, the 'Size', 'Nodes', 'pgr', and 'er' columns exhibit a significant number of outlier values.
